# Задача 1. Биномиальное распределение

**Условие**  
Вероятность того, что лампа перегорит в течение гарантийного срока, равна $ p = 0{,}1 $.  
Проверяется партия из $ n = 8 $ ламп.  
Случайная величина $ X $ — число перегоревших ламп.

**Требуется**  
а) составить закон распределения $ X $;  
б) найти $ \mathbb{E}[X] $, $ \mathrm{D}[X] $, $ \sigma[X] $, моду;  
в) построить многоугольник распределения.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 8, 0.1
X = np.arange(0, n + 1)
dist = binom(n, p)
prob = dist.pmf(X)

print("x:", " ".join(f"{x:2d}" for x in X))
print("p:", " ".join(f"{pi:6.4f}" for pi in prob))

mean = dist.mean()
var = dist.var()
std = np.sqrt(var)

max_p = prob.max()
mode = [int(x) for x in X if abs(prob[x] - max_p) < 1e-12]

print(f"\nM[X] = {mean:.4f}")
print(f"D[X] = {var:.4f}")
print(f"σ[X] = {std:.4f}")
print(f"Mo[X] = {mode}")

plt.figure(figsize=(7, 4))
plt.plot(X, prob, 'bo-', linewidth=2, markersize=7)
for x, pi in zip(X, prob):
    if pi > 0.001:
        plt.text(x, pi + 0.005, f'{pi:.3f}', ha='center', fontsize=8, color='darkblue')
plt.title('Многоугольник распределения: X ~ Binomial(n=8, p=0.1)')
plt.xlabel('x — число перегоревших ламп')
plt.ylabel('P(X = x)')
plt.xticks(X)
plt.ylim(0, max(prob) * 1.1)
plt.grid(True, alpha=0.6, linestyle='--')
plt.tight_layout()
plt.show()
```
