# Задача 2. Геометрическое распределение

**Условие**  
Вероятность обнаружения дефекта при проверке изделия равна $ p = 0{,}1 $.  
Проверка производится до первого обнаружения дефекта.  
Случайная величина $ X $ — число проверенных изделий.

**Требуется**  
а) составить закон распределения $ X $;  
б) найти $ \mathbb{E}[X] $, $ \mathrm{D}[X] $, $ \sigma[X] $, моду;  
в) определить $ \mathbb{P}(X \geq 6) $;  
г) построить многоугольник распределения.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p = 0.1
K = np.arange(1, 11)
dist = geom(p)
prob = dist.pmf(K)

print("k:", " ".join(f"{k:2d}" for k in K))
print("p:", " ".join(f"{pk:6.4f}" for pk in prob))

mean = dist.mean()
var = dist.var()
std = np.sqrt(var)
mode = [1]

print(f"\nM[X] = {mean:.4f}")
print(f"D[X] = {var:.4f}")
print(f"σ[X] = {std:.4f}")
print(f"Mo[X] = {mode}")

p_ge_6 = (1 - p) ** 5
print(f"\nP(X ≥ 6) = {p_ge_6:.6f}")

plt.figure(figsize=(7, 4))
plt.plot(K, prob, 'ro-', linewidth=2, markersize=7)
for k, pk in zip(K, prob):
    if pk > 0.001:
        plt.text(k, pk + 0.003, f'{pk:.3f}', ha='center', fontsize=8, color='darkred')
plt.title('Многоугольник распределения: X ~ Geometric(p=0.1)')
plt.xlabel('k — число проверенных изделий')
plt.ylabel('P(X = k)')
plt.xticks(K)
plt.ylim(0, max(prob) * 1.1)
plt.grid(True, alpha=0.6, linestyle='--')
plt.tight_layout()
plt.show()
```
